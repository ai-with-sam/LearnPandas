```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)

data = {
    'order_id': ['O1','O2','O3','O4','O5','O6','O7','O8','O9','O10','O11','O12'],
    'city': ['Mumbai','Delhi','Mumbai','Bangalore','Delhi','Mumbai','Delhi',
             'Bangalore','Mumbai','Delhi','Bangalore','Mumbai'],
    'category': ['Tech','Furniture','Tech','Office','Tech','Furniture','Office',
                 'Tech','Office','Furniture','Furniture','Tech'],
    'sales': [25000, 8000, 15000, 3000, 32000,np.nan , 5500, 18000, 4200, 9800, 7600, 21000],
    'profit': [5000, -500, 3200, 800, 8000, 1200, 600, 4000, -300, 1900, 1100, 5200],
    'order_date': pd.to_datetime(['2025-01-05','2025-01-08','2025-02-10','2025-02-12',
                                   '2025-02-15','2025-03-01','2025-03-05','2025-03-10',
                                   '2025-03-15','2025-04-01','2025-04-05','2025-04-10'])
}
df = pd.DataFrame(data)
df
```

```python
df.pivot_table(
    index='city',
    values='sales',
    columns='category',
    aggfunc='sum'
    )
```

groupby() vs pivot_table()

groupby()-Used for analysis.
pivot_table()-Used for reports.

```python
# Pivot Table with Multiple Values

df.pivot_table(
    index='city',
    values=['sales','profit'],
    columns='category',
    aggfunc=['sum','mean'],
    fill_value=0 # Replace Nan values to 0
    )
```

Margins (Grand Totals)
-Add totals to the report.



```python
df.pivot_table(
    index='city',
    values='sales',
    columns='category',
    aggfunc='sum',
    margins=True
    )
```

```python
df.pivot_table(
    index='city',
    values='sales',
    columns='category',
    aggfunc=['sum','mean'],
    margins=True
    )
```

```python
df.pivot_table(
    index='city',
    values=['sales','profit'],
    columns='category',
    aggfunc='sum',
    margins=True
    )
```

```python
df.pivot_table(
    index='city',
    values=['sales','profit'],
    columns='category',
    aggfunc='sum',
    margins=True,
    margins_name='Total_sales' # Adds label to Total values
    )
```

```python
pivotlearn = df.pivot_table(
    index='city',
    values=['sales','profit'],
    columns='category',
    aggfunc='sum',
    margins=False, # Removes the summed up column
    margins_name='Total_sales' # Adds label to Total values
    )
pivotlearn
```

```python
pivotlearn.loc['Average'] = pivotlearn.mean() #Calculates the mean of each column
                                              # Adds a new row named "Average" to the DataFrame
pivotlearn
```

```python
pivotlearn['Average'] = pivotlearn.mean(axis=1)
# This selects (or creates) a column named "Average".
#Since it doesn't exist, pandas creates it.

pivotlearn
```
