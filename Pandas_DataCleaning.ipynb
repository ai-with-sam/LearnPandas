```python
import pandas as pd

data = {
    "Order_ID": [1001, 1002, 1002, 1004],
    "Customer": ["Rahul", "Priya", "Priya", "Vijay"],
    "Age": [25, None, None, "twenty-eight"],
    "Order_Value": [500, 1200, 1200, None]
}

df3 = pd.DataFrame(data)
print(df3)
```

```python
df3.isnull()
```

```python
df3.isnull().sum()
```

```python
df3
```

```python
df3.duplicated()
```

```python
df3.duplicated().sum()
```

```python
df3.drop_duplicates()
```

```python
df3["Age"] = df3["Age"].replace("twenty-eight", 28)
df3["Age"] = pd.to_numeric(df3["Age"])
df3['Age']
```
